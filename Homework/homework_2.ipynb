#### Setup: Load libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

#### Task 1: Import and load the dataset

```python
dataset = load_breast_cancer()

print(dataset.data)
print(dataset.data.shape)

df = pd.DataFrame(dataset.data)
df.columns = dataset.feature_names
df["TARGET"] = dataset.target

print(df.head())
```

#### Task 2: Build a correlation heatmap to check for each feature's correlation with the labels

```python
f, ax = plt.subplots(figsize=(20, 15))
sns.heatmap(df.corr().abs(), cmap="Blues", annot=True)
```

#### Task 3: Build a correlation heatmap to check for multicollinearity between features

```python
f, ax = plt.subplots(figsize=(20, 15))
sns.heatmap(df.corr().abs(), cmap="Blues", annot=True, vmin=-1, vmax=1, center=0, mask=np.triu(df.corr()))
```

#### Task 4: Scale and transform the data using a StandardScaler() object

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
print(df_scaled)
```

#### Tasks 5-8

```python
# Helper function
def model(data, target): 
    X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.25, random_state=0)
    pipeline = make_pipeline(SVC(gamma="scale"))
    model = pipeline.fit(X_train, y_train)
    return (X_test, y_test, model)
```

```python
X_test, y_test, model = model(df, df["TARGET"])
prediction = model.predict(X_test)
print("MSE", mean_squared_error(y_test, prediction))
print("R-Squared", r2_score(y_test, prediction))
```
