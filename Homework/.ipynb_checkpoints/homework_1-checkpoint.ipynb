### Homework 1: Linear Regression

#### Tasks:
    - Load the dataset,
    - Perform basic exploratory data analysis,
    - Create a correlation heatmap,
    - Train the model based on the train set,
    - Make predictions on the test set and compare to the actual targets,
    - Compute MSE and R-Squared score of the model.

#### Setup: Load libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

#### Task 1: Load the dataset

```python
boston = load_boston()  # Loads the dataset we will be using for this notebook

print(boston.data)  # Prints out the data from the dataset
print(boston.data.shape)  # Prints the size of the dataset (length, width)

bos = pd.DataFrame(boston.data)  # Creates a dataframe using the data from the dataset
bos.columns = boston.feature_names  # Stores the columns from the dataset in a variable
bos["PRICE"] = boston.target  # Sets the target we will be using for this notebook

print(bos.head())  # Returns the first five rows of the dataset
```

#### Task 2: Perform basic exploratory data analysis

```python
print(bos.describe())  # Returns general statistics of the dataset (in thousands)
```

```python
f, ax = plt.subplots(ncols=7, nrows=2, figsize=(12, 9))  # Sets the size of the iterated boxplot area
ax = ax.flatten()  # Stores amount of axes into an array
i = 0  # Index
for k, v in bos.items():
    sns.boxplot(y=k, data=bos, palette="Blues", ax=ax[i])  # Creates a boxplot after each iteration
    i += 1  # Increment the index
plt.tight_layout(pad=1.5)  # Since multiple plots are being shown, adds padding around each plot
```

```python
bos = bos[~(bos["PRICE"] >= 50.0)]  # Removes outliers from price for data analysis purposes
```

```python
f, ax = plt.subplots(ncols=7, nrows=2, figsize=(12, 6))  # Sets the size of the iterated distribution plot area
ax = ax.flatten()  # Stores amount of axes into an array
i = 0  # Index
for k, v in bos.items():
    sns.distplot(v, ax=ax[i])  # Creates a distribution plot after each iteration
    i += 1  # Increment the index
plt.tight_layout(pad=1.5)  # Since multiple plots are being shown, adds padding around each plot
```

#### Task 3: Create a correlation heatmap

```python
f, ax = plt.subplots(figsize=(12, 9))  # Sets the size of the heatmap
sns.heatmap(bos.corr().abs(), cmap="Blues", annot=True)  # Returns a Pearson correlation heatmap with absolute values
```

#### Task 4: Train the model based on the train set

```python
bos = pd.DataFrame(boston.data)  # Reinitialize the dataframe since it was modified
bos["PRICE"] = boston.target  # Reinitialize the target since it was also modified

X = bos.drop("PRICE", axis=1)  # Removes target from independent variables
y = bos["PRICE"]  # Sets target variable to price

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
lm = LinearRegression()  # Creates the linear regressor
lm.fit(X_train, y_train)  # Trains the model using the train set
```

#### Task 5: Make predictions on the test set and compare to the actual targets

##### Train set predictions

```python
y_pred = lm.predict(X_train)
print(y_pred[:10])  # Prints first ten values for easier visualization
```

##### Test set predictions

```python
y_test_pred = lm.predict(X_test)
print(y_test_pred[:10])  # Prints first ten values for easier visualization
```

##### Prediction visualization

```python
f, ax = plt.subplots(figsize=(12, 9))  # Sets the size of the scatter plot
sns.scatterplot(x=y_train, y=y_pred, palette="Blues", alpha=0.5)  # Returns scatter plot
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
```

#### Task 6: Compute MSE and R-Squared score of the model

##### Train data scores

```python
print("MSE", mean_squared_error(y_train, y_pred))
print("R-Squared", r2_score(y_train, y_pred))
```

##### Test data scores

```python
print("MSE", mean_squared_error(y_test, y_test_pred))
print("R-Squared", r2_score(y_test, y_test_pred))
```
